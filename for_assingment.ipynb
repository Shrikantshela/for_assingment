```python
#Q3. Create a python program to compute the electricity bill for a household.
The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
ans:-
# Input the total units consumed
units_consumed = int(input("Enter the total units consumed in a month: "))

# Initialize variables to keep track of the total bill
total_bill = 0

# Calculate the bill based on the given per-unit charges
if units_consumed <= 100:
    total_bill = units_consumed * 4.5
elif units_consumed <= 200:
    total_bill = 100 * 4.5 + (units_consumed - 100) * 6
elif units_consumed <= 300:
    total_bill = 100 * 4.5 + 100 * 6 + (units_consumed - 200) * 10
else:
    total_bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units_consumed - 300) * 20

# Print the electricity bill
print(f"Electricity Bill: Rs. {total_bill:.2f}")
```

```python
#Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.
ans:-for using for loop:-
# Using a for loop to generate the list
cubes_divisible_by_4_or_5 = []

for num in range(1, 101):
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_divisible_by_4_or_5.append(num)

# Print the list
print("Numbers whose cubes are divisible by 4 or 5:")
print(cubes_divisible_by_4_or_5)
# Using a while loop to generate the list
cubes_divisible_by_4_or_5 = []
num = 1

while num <= 100:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubes_divisible_by_4_or_5.append(num)
    num += 1

# Print the list
print("Numbers whose cubes are divisible by 4 or 5:")
print(cubes_divisible_by_4_or_5)
```

```python
#Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"
ans:-
# Input string
string = "I want to become a data scientist"

# Define a function to count vowels
def count_vowels(input_string):
    # Initialize a counter for vowels
    vowel_count = 0

    # Define a set of vowels to check against (both lowercase and uppercase)
    vowels = set("aeiouAEIOU")

    # Iterate through the characters in the input string
    for char in input_string:
        if char in vowels:
            vowel_count += 1

    return vowel_count

# Call the function to count vowels in the given string
vowel_count = count_vowels(string)

# Print the result
print(f"Number of vowels in the string: {vowel_count}")
This program defines a function count_vowels that iterates through each character in the input string and checks if it is a vowel (both lowercase and uppercase). If a character is a vowel, it increments the vowel_count variable. Finally, it prints the total count of vowels in the string.

When you run this program with the given string, it will count the vowels and print the result.
```
